# Import required library (pandas) for Handling Json File

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


```

#Reading the json file and formatting its orientation as columns 

```python
df=pd.read_json('/content/drive/MyDrive/Vishesh kushwaha_revedor_python_assignment/data.json',orient='columns',lines=True)
df.head()
```

#Creating a new dataframe to rename the coulmns

```python
new_df=pd.DataFrame(df)
new_df.rename(columns={"a":"address","liid":"linkedin_id","n":"name","t":"mobile_num","e":"email","linkedin":"profile_url"},inplace=True)
new_df.head(3)
```

#Creating csv files

```python
file='/content/drive/MyDrive/Vishesh kushwaha_revedor_python_assignment'
```

```python
new_df.to_csv(file+'data_main.csv',columns=['name','linkedin_id','address','profile_url'],sep=",")
```

```python
new_df.to_csv(file+'data_email.csv',columns=['linkedin_id','email'],sep=",")
```

```python
new_df.to_csv(file+'data_number.csv',columns=['linkedin_id','mobile_num'],sep=",")
```

##Analysis of Data

```python
temp=pd.read_csv(file+'data_main.csv').drop(columns=['Unnamed: 0'])
temp.info()
```

```python
x=temp.address.isna().sum()
address_na_percent=x/len(temp)*100
address_nonna_percent=(1-(x/len(temp)))*100
print('Address null %')
labels=['NULL '+str(address_na_percent)+"%",'NON-NULL '+str(address_nonna_percent)+"%"]
sizes=[address_na_percent,address_nonna_percent]
colours=['red','green']
plt.pie(sizes,labels=labels,colors=colours)
plt.tight_layout()
plt.axis('equal')
plt.show()
```

```python
temp2=pd.read_csv(file+'data_email.csv').drop(columns=['Unnamed: 0'])
temp2.info()
```

```python
x1=temp2.email.isna().sum()
email_na_percent=x1/len(temp2)*100
email_nonna_percent=(1-(x1/len(temp2)))*100
print('email null %')
labels=['NULL '+str(email_na_percent)+"%",'NON-NULL '+str(email_nonna_percent)+"%"]
sizes=[email_na_percent,email_nonna_percent]
colours=['red','green']
plt.pie(sizes,labels=labels,colors=colours)
plt.tight_layout()
plt.axis('equal')
plt.show()
```

```python
temp3=pd.read_csv(file+'data_number.csv').drop(columns=['Unnamed: 0'])
temp3.info()
```

```python
x2=temp3.mobile_num.isna().sum()
number_na_percent=x2/len(temp2)*100
number_nonna_percent=(1-(x2/len(temp2)))*100
print('email null %')
labels=['NULL '+str(number_na_percent)+"%",'NON-NULL '+str(number_nonna_percent)+"%"]
sizes=[number_na_percent,number_nonna_percent]
colours=['red','green']
plt.pie(sizes,labels=labels,colors=colours)
plt.tight_layout()
plt.axis('equal')
plt.show()
```
